```python
# Ejemplo para leer las imágenes.

# Índice aleatorio
idx = np.random.randint(0,len(img_paths))

# Leer fotografía de células y label
cell = cv2.imread(img_paths[idx], cv2.COLOR_BGR2RGB)
cell = cv2.cvtColor(cell, cv2.COLOR_BGR2RGB)

xdim = 256
ydim = 256

xscale = xdim/cell.shape[0]
yscale = ydim/cell.shape[1]

cell = cv2.resize(cell, (xdim, ydim), interpolation = cv2.INTER_CUBIC)


fig, ax = plt.subplots()
im = ax.imshow(cell)

for coords in xys[idx]:
    ax.plot(xscale*coords[0], yscale*coords[1] ,marker = "o", markersize = 5, color='green')
plt.show()

fig, ax = plt.subplots()
im = ax.imshow(cell)

print(img_paths[idx], cell.shape, spcs_labels[idx])
```

```python
from scipy.ndimage import gaussian_filter
pmap = np.zeros([ydim,xdim]).astype("uint8")

#Si el punto se encuentra dentro de la imagen, entonces se asigna un 1 en la posición x,y del label.
for coords in xys[idx]:
    x = xscale*coords[0]
    y = yscale*coords[1]
    if x <= xdim and y <= ydim:
        pmap[int(y)][int(x)] = 1

dmap = gaussian_filter(pmap, sigma=(1.5, 1.5), order=0)
```

```python
plt.imshow(pmap, cmap='hot', interpolation='nearest')
plt.title("Mapa de puntos")
plt.show()


plt.imshow(dmap, cmap='hot', interpolation='nearest')
plt.title("Mapa de densidad")
plt.show()

print(
np.sum(pmap),
np.round(np.sum(dmap),0),
len(xys[idx]))
```
